```python
#Função para eliminar texto de número
def keep_numbers(dataframe, columns):
    for column_name in columns:
        dataframe[column_name] = pd.to_numeric(dataframe[column_name], errors='coerce')
    return dataframe

numeric_columns = ['age', 'annual_income', 'monthly_inhand_salary', 'num_bank_accounts', 'num_credit_card',
                    'interest_rate', 'num_of_loan', 'delay_from_due_date', 'num_of_delayed_payment', 'changed_credit_limit',
                    'num_credit_inquiries', 'outstanding_debt', 'credit_utilization_ratio', 'total_emi_per_month',
                    'amount_invested_monthly', 'monthly_balance']

keep_numbers(df_credito, numeric_columns)
```

## Tratar colunas uma a uma

### Para as numéricas:

1.Limitar valores das colunas que não fazem sentido, como idade, número de contas bancárias e número de cartões de crédito

2.Substituir os valores NaN das pela mediana de cada coluna


### Para as de texto

Cada uma terá o tratamento adequado mais adiante

```python
#Função para limitar valores nas colunas

def limit_values(dataframe, column_name, min_number, max_number):
    dataframe.loc[(dataframe[column_name] <= min_number) | (dataframe[column_name] >= max_number), column_name] = None
    return dataframe
```

```python
#Limitar valores das colunas 
#age -> de 16 a 100
limit_values(df_credito,'age', 16, 100)
```

```python
#Bank Accounts -> de 0 a 100
limit_values(df_credito,'num_bank_accounts', 0, 100)
```

```python
#Credit Card -> de 0 a 100
limit_values(df_credito,'num_credit_card', 0, 100)
```

```python
#num_of_loan -> de 0 a 100
limit_values(df_credito,'num_of_loan', 0, 100)
```

```python
#Substituir todos os valores vazios das colunas numéricas pela mediana
def replace_nan_with_median(dataframe):
    numeric_columns = dataframe.select_dtypes(include='number').columns
    
    for column_name in numeric_columns:
        median_value = dataframe[column_name].median()
        dataframe[column_name].fillna(median_value, inplace=True)
    
    return dataframe

replace_nan_with_median(df_credito)
```

## Colunas de Texto (Categóricas)

```python
#Coluna occupation - substituindo _______ por Unknown

df_credito['occupation'] = df_credito['occupation'].replace('_______', 'Unknown')
```

```python
# _______ foi eliminado e substituído por Unknown como podemos ver abaixo
occupations = df_credito['occupation'].unique()
occupations
```

```python
#Contando quanto temos de cada profissão -> vemos que Unknown é o que mais aparece

occupations_values = df_credito['occupation'].value_counts()
occupations_values
```

```python
# Separando Type of loan em colunas independentes e binárias
df_credito['type_of_loan'] = df_credito['type_of_loan'].str.replace(', ', ',')
df_credito['type_of_loan'] = df_credito['type_of_loan'].str.replace('and ', '')
loan_dummies = df_credito['type_of_loan'].str.get_dummies(',')
df_credito = pd.concat([df_credito, loan_dummies], axis=1)
df_credito = df_credito.drop(columns=['type_of_loan'])
credit_score_column = df_credito.pop('credit_score')
df_credito['credit_score'] = credit_score_column
df_credito
```

```python
# Transformando Credit History Age em anos em três etapas

# 1. Substituir todos os NA por NaN

df_credito['credit_history_age'].replace('NA', np.nan, inplace=True)

#2. Substituir texto por número

import re

def convert_to_years(value):
    if pd.notna(value):
        match = re.match(r'(\d+) years and (\d+) months', value)
        if match:
            years = int(match.group(1))
            months = int(match.group(2))
            return years + months / 12
    return value

df_credito['credit_history_age'] = df_credito['credit_history_age'].apply(convert_to_years)

#3. Substituir os Nan por mediana
replace_nan_with_median(df_credito)
df_credito
```
